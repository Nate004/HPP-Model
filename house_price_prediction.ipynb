## Test train split
1. Training set
2. Test set
#### Methods
1. Validation set approach: random division of data into 2 groups usually 80:20. to be used in the case of large no. of observations
2. Leave one out cross validation: leave one observation everytime from training set
3. K-fold validation - divides the data to k set, one for testing and k - 1 for training

### Bias-variance trade-off
1. Expected test error E(e) = E(bias) + E(var) + E(error)
      E(error) = variance of error, irreducible
      E(var) = amount by which predicted function will change if we change training dataset
      E(bias) = error due to approximation of complex relationship as a simpler model such as linera model
2. Overfitting: this is when we use the model function to meet more data points(more flexibility of curve) reduces training error but increases test error. 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size = 0.2, random_state = 0) #to create train and test variables
```

```python
from sklearn import preprocessing  ##this is for standardization of independent variable in ridge regression
```

```python
scalar = preprocessing.StandardScaler().fit(X_train)  ## to scale the x parameter
```

```python
X_train_s = scalar.transform(X_train)
```

```python
X_test_s = scalar.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)
lm_r.fit(X_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(X_test_s))
```

### To find the value of alpha (giving multiple values) that gives best optimization, use validation curve function from sklearn

```python
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
```

```python
param_range = np.logspace(-2,8,100)    ## to create log values btw 10^-2 to 10^8
```

```python
param_range
```

```python
train_scores, test_scores = validation_curve(Ridge(), X = X_train_s, y = y_train, param_name =  "alpha",param_range = param_range, scoring = 'r2')
```

```python
print(train_scores)
print(test_scores)
```

```python
train_mean = np.mean(train_scores, axis = 1)
```

```python
test_mean = np.mean(test_scores, axis = 1)
```
